## Part 1 - Read the 2021 course schedule into a DataFrame

```python
from bs4 import BeautifulSoup
```

```python
import re
```

```python
dom = BeautifulSoup(open('./nyu_cs_course_schedule.htm'))
```

```python
lis = dom.select('.schedule-listing li')
```

```python
def id(li): 
    spans = li.select('span')
    return spans[0].get_text().strip().split('\n')[0].upper()
def title(li): 
    spans = li.select('span')
    return spans[1].get_text().strip()

def instructor(li): 
    spans = li.select('span')
    return spans[2].get_text().strip().split('\n')[0]

def time(li): 
    spans = li.select('span')
    return spans[3].get_text().strip().split('\n')[0]

tmp = [(id(li), title(li), instructor(li), time(li)) for li in lis]
tmp[:10]
```

```python
import pandas as pd
schedule_df = pd.DataFrame(tmp, columns = ['Number-Section', 'Name', 'Instructor', 'Time'])
```

```python
schedule_df
```

## Part 2: Break apart 'Number-Section' column into two separate columns: 'Number' and 'Section'

```python
schedule_df['Number'] = schedule_df['Number-Section'].str.split('-').str[:2].str.join('-')
```

```python
schedule_df
```

```python
schedule_df['Section'] = schedule_df['Number-Section'].str.split('-').str[-1:].str.join('')
```

## Part 4: Read the 2021 course catalog into a DataFrame

```python
dom = BeautifulSoup(open('./nyu_cs_course_catalog.htm'))
```

```python
import re
def numbers(li): 
    spans = li.select('p')
    return spans[0].get_text().strip().split(' ')[0].upper()
    
def points(li): 
    content = li.select('p')[1].get_text()
    m = re.search('([0-9 -]+) Points', content)
    if m: 
        return m[1].strip()
def prereqs(li): 
    content = li.select('p')[2].get_text()
    content_display = content.partition("Prerequisites: ")[2]
    return content_display
    
courses = dom.select('ul.courses-listing > li')
tmp_courses = [(numbers(course), prereqs(course), points(course)) for course in courses]
```

```python
catalog_df = pd.DataFrame(tmp_courses, columns=['Number', 'Prereqs', 'Points'])
```

## Part 6: Combine both DataFrames

```python
res = pd.merge(schedule_df, catalog_df, on='Number', how='left')
```

```python
res[['Number', 'Name', 'Instructor', 'Time', 'Prereqs', 'Points']]
```
